```python
# Pada dataset yang akan digunakan, hanya dipilih feature uhh-PDRB
X_pca = df. iloc[:, 2:]
X_pca.head(3)
```

```python
# Proses scaling data pada variable X_pca untuk dilakukan PCA
scaler = StandardScaler()
data =  scaler.fit_transform(X_pca)
data
```

```python
# Fungsi untuk menghitung nilai eigen
def compute_eigenvalues(data):
    covariance_matrix = np.cov(data.T)
    eigenvalues, _ = np.linalg.eig(covariance_matrix)
    return eigenvalues

# Hitung nilai eigen
eigenvalues = compute_eigenvalues(data)
eigenvalues
```

```python
# Proses PCA, membagi menjadi n_components yang diinginkan sesuai dgn eigenvalue
from sklearn.decomposition import PCA
pca = PCA(n_components=2)  # Specify the number of components you want to keep
principal_components = pca.fit_transform(data)

# Hasil PCA disusun dalam dataframe
df_pca = pd.DataFrame(principal_components)
df_pca
```

```python
# Rename kolom PCA yang sudah terbentuk
df_pca = df_pca.rename(
    columns={0: 'PC1',
             1: 'PC2'}
)
df_pca
```

# Clustering

## K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# Normalization
scaler = StandardScaler()
X_normalized = scaler.fit_transform(df_pca)

# Calculate silhouette score for different number of clusters
silhouette_scores = []
for n_clusters in range(2, 7):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels_kmeans = kmeans.fit_predict(X_normalized)
    silhouette_avg = silhouette_score(X_normalized, labels_kmeans)
    silhouette_scores.append(silhouette_avg)

# Elbow Method: Plotting the silhouette scores
plt.figure(figsize=(10, 5))
plt.plot(range(2, 7), silhouette_scores, marker='o', linestyle='-')
plt.title('Silhouette Score vs. Number of Clusters (K-Means)')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(np.arange(2, 7))
plt.grid(True)
plt.show()

# Using the Elbow Method to determine the optimal number of clusters
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + 2  # Adding 2 to account for starting from 2 clusters
print("Optimal number of clusters (by silhouette score):", optimal_num_clusters)

# Performing k-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_num_clusters, random_state=42)
labels_kmeans = kmeans.fit_predict(X_normalized)

# Calculate Silhouette Score with optimal number of clusters
silhouette_avg = silhouette_score(X_normalized, labels_kmeans)
print("Silhouette Score with optimal number of clusters:", silhouette_avg)

# Visualization
plt.scatter(df_pca.iloc[:, 0], df_pca.iloc[:, 1], c=labels_kmeans, cmap='viridis', marker='o')

# Annotate each point with its name
for i, label in enumerate(labels_kmeans):
    plt.annotate(df['A'][i], (df_pca.iloc[i, 0], df_pca.iloc[i, 1]),
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8)

plt.title(f'K-Means Clustering (Silhouette Score: {silhouette_avg:.2f})')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.axhline(y=0)
plt.axvline(x=0)
plt.show()
```

## Spectral Clustering

```python
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# Normalization
scaler = StandardScaler()
X_normalized = scaler.fit_transform(df_pca)

# Calculate silhouette score for different number of clusters
silhouette_scores = []
for n_clusters in range(2, 7):
    spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
    labels_spectral = spectral_clustering.fit_predict(X_normalized)
    silhouette_avg = silhouette_score(X_normalized, labels_spectral)
    silhouette_scores.append(silhouette_avg)

# Elbow Method: Plotting the silhouette scores
plt.figure(figsize=(10, 5))
plt.plot(range(2, 7), silhouette_scores, marker='o', linestyle='-')
plt.title('Silhouette Score vs. Number of Clusters (Spectral Clustering)')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(np.arange(2, 7))
plt.grid(True)
plt.show()

# Using the Elbow Method to determine the optimal number of clusters
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + 2  # Adding 2 to account for starting from 2 clusters
print("Optimal number of clusters (by silhouette score):", optimal_num_clusters)

# Performing spectral clustering with the optimal number of clusters
spectral_clustering = SpectralClustering(n_clusters=optimal_num_clusters, affinity='nearest_neighbors', random_state=42)
labels_spectral = spectral_clustering.fit_predict(X_normalized)

# Calculate Silhouette Score with optimal number of clusters
silhouette_avg = silhouette_score(X_normalized, labels_spectral)
print("Silhouette Score with optimal number of clusters:", silhouette_avg)

# Visualization
plt.scatter(df_pca.iloc[:, 0], df_pca.iloc[:, 1], c=labels_spectral, cmap='viridis', marker='o')

# Annotate each point with its name
for i, label in enumerate(labels_spectral):
    plt.annotate(df['A'][i], (df_pca.iloc[i, 0], df_pca.iloc[i, 1]),
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8)

plt.title(f'Spectral Clustering (Silhouette Score: {silhouette_avg:.2f})')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.axhline(y=0)
plt.axvline(x=0)
plt.show()
```

## DBSCAN

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Normalization
scaler = StandardScaler()
X_normalized = scaler.fit_transform(df_pca)

# Parameter ranges for grid search
eps_range = np.linspace(0.1, 0.5, 4)  # Adjust range as needed
min_samples_range = range(2, 10)  # Adjust range as needed

# Store silhouette scores
silhouette_scores = []

# Grid search for optimal parameters
for eps in eps_range:
    for min_samples in min_samples_range:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_normalized)
        unique_labels = np.unique(labels)
        if len(unique_labels) > 1:  # Calculate silhouette score only if more than one cluster
            silhouette_avg = silhouette_score(X_normalized, labels)
            silhouette_scores.append((eps, min_samples, silhouette_avg))

# Plotting elbow graph
silhouette_scores = np.array(silhouette_scores)
plt.figure(figsize=(10, 6))
plt.plot(silhouette_scores[:, 0], silhouette_scores[:, 2], marker='o', linestyle='-')
plt.title('Elbow Method for DBSCAN Parameter Optimization')
plt.xlabel('Eps')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Choose parameters with the highest silhouette score
best_params = silhouette_scores[silhouette_scores[:, 2].argmax()]
best_eps, best_min_samples = best_params[0], int(best_params[1])

# Perform clustering with the best parameters
dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
labels_dbscan = dbscan.fit_predict(X_normalized)

# Calculate silhouette score
silhouette_avg = silhouette_score(X_normalized, labels_dbscan)

# Visualize clustering results
plt.scatter(df_pca.iloc[:, 0], df_pca.iloc[:, 1], c=labels_dbscan, cmap='viridis', marker='o')
for i, label in enumerate(labels_dbscan):
    plt.annotate(df['A'][i], (df_pca.iloc[i, 0], df_pca.iloc[i, 1]),
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8)
plt.title(f'DBSCAN Clustering (Silhouette Score: {silhouette_avg:.2f}, eps={best_eps}, min_samples={best_min_samples})')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.axhline(y=0)
plt.axvline(x=0)
plt.show()
```
